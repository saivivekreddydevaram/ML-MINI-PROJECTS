```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
path = 'https://raw.githubusercontent.com/saivivekreddydevaram/ML-PROJECTS/refs/heads/main/NaiveBayesSpamClassifier/spam_ham_dataset.csv'
mail_data = pd.read_csv(path)
```

```python
mail_data.head(20)
```

```python
def clean_text(text):
  text = text.lower()
  text = re.sub(r'\d+','',text)
  text = re.sub(r'[^\w\s]','',text)
  text = re.sub(r'[\n\r\\]+',' ',text)
  text = re.sub(r'\s+',' ',text).strip()
  text = text.split()
  return text
```

```python
mail_data.drop(columns= 'label',axis = 1,inplace = True)
```

```python
split_index = int(len(mail_data)*0.8)

train_data = mail_data[:split_index]
test_data = mail_data[split_index:]

X_train = train_data['text'].apply(clean_text)
Y_train = train_data['label_num']

X_test = test_data['text'].apply(clean_text)
Y_test = test_data['label_num']
```

```python
all_words = []
for message in X_train:
  all_words.extend(message[:])
vocab = list(set(sorted(all_words)))
vocab_dict = {}
for i in range(0,len(vocab)):
  vocab_dict[vocab[i]] = i + 1
print(vocab_dict)
```

```python
siz = len(vocab_dict)
count_words_y_1 = np.zeros(siz)
count_words_y_0 = np.zeros(siz)
total_words_0 = 0
total_words_1 = 0
for email,label in zip(X_train,Y_train):
  for word in email:
    if word in vocab_dict:
      index = vocab_dict[word]
      if label == 1:
        count_words_y_1[index] += 1
        total_words_1 += 1
      else:
        count_words_y_0[index] += 1
        total_words_0 += 1
```
